## Load data

```python
import numpy as np

filename = 'avila_250_measures-1M.npz'
data = fl.load_generated_data(os.path.join(PROJ_DIR, 'results', filename))
```

```python
qmean_mean = np.mean(data['qmeans'])
qmean_std = np.std(data['qmeans'])

z_indices = (data['qmeans'] - qmean_mean) / qmean_std
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['qmeans'], data['qmean_invs'], s=1)
plt.show()
plt.scatter(data['qmeans'], data['qmean_bas'], s=1)
plt.show()
plt.scatter(z_indices, data['qmeans'], s=1)
plt.show()
plt.scatter(z_indices, data['qmean_bas'], s=1)
plt.show()
```

```python
best_index_faithfulness_correlations = np.argmax(data['faithfulness_correlations'])
best_index_monotonicity_correlations = np.argmax(data['monotonicity_correlations'])
best_index_qmean_invs = np.argmax(data['qmean_invs'])
print(best_index_faithfulness_correlations)
print(best_index_monotonicity_correlations)
print(best_index_qmean_invs)
```

```python
z_index_best_faithfulness_correlation  = (data['qmeans'][best_index_faithfulness_correlations] - qmean_mean) / qmean_std
z_index_best_monotonicity_correlation  = (data['qmeans'][best_index_monotonicity_correlations] - qmean_mean) / qmean_std
z_index_best_qmean_inv  = (data['qmeans'][best_index_qmean_invs] - qmean_mean) / qmean_std

print(z_index_best_faithfulness_correlation)
print(z_index_best_monotonicity_correlation)
print(z_index_best_qmean_inv)
```

```python
print(data['rankings'][np.argmax(data['faithfulness_correlations'])])
print(data['rankings'][np.argmax(data['monotonicity_correlations'])])
print(data['rankings'][np.argmax(data['qmeans'])])
print(data['rankings'][np.argmax(data['qmean_invs'])])
```

```python
def attribution_to_ranking(attr):
    return list(map(lambda x:x[1], sorted(list(map(lambda x: (x[1],x[0]),enumerate(attr))))))

print(attribution_to_ranking(data['rankings'][np.argmax(data['faithfulness_correlations'])]))
print(attribution_to_ranking(data['rankings'][np.argmax(data['monotonicity_correlations'])]))
print(attribution_to_ranking(data['rankings'][np.argmax(data['qmeans'])]))
print(attribution_to_ranking(data['rankings'][np.argmax(data['qmean_invs'])]))
```

```python
top10_faithfulness_correlations = np.argpartition(data['faithfulness_correlations'], -10)[-10:]
top10_monotonicity_correlations = np.argpartition(data['monotonicity_correlations'], -10)[-10:]
top10_qmeans = np.argpartition(data['qmeans'].flatten(), -10)[-10:]
top10_qmean_invs = np.argpartition(data['qmean_invs'].flatten(), -10)[-10:]
print(top10_faithfulness_correlations)
print(top10_monotonicity_correlations)
print(top10_qmeans)
print(top10_qmean_invs)
```
